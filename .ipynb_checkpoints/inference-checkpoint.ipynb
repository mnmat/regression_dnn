```python
import torch
from utils.datasets import *
from models.models import *
from utils.validation import *
```

```python
# Create Dataset

root = "/Users/markmatthewman/Projects/PhD/EgammaDNN/data/GenSim/TICLv4_Mustache/electron"
fname = "HLTAnalyzerTree_IDEAL_Flat_test.root"
tree = "egRegDataEcalHLTV1"
batches = 256

dataset = RegressionDataset(root,fname,tree)
dataloader = DataLoader(dataset,batch_size=batches,shuffle=True)
```

```python
# Load Model

model_path = "saved_models/v1"
model_name = "v1"
model = NeuralNetwork()
device = torch.device("cpu")

if model_path is not None:
    checkpoint = torch.load(os.path.join(model_path,model_name))
    model.load_state_dict(checkpoint["model_state_dict"])
model.to(device)
model.eval()
```

```python
# Validation loop + fill histograms

def writeHistograms(X,prediction,y,hists):
    hists["h_reco_tot"].fill(X[FEATURES["rawEnergy"]])
    hists["h_pred_tot"].fill(prediction)
    hists["h_cp"].fill(y)
    hists["h_reco_tot_over_cp"].fill(X[FEATURES["rawEnergy"]]/y)
    hists["h_pred_tot_over_cp"].fill(prediction / y)
    return hists

hists = createHists()

for val_batch in dataloader:
    inpt = val_batch["features"].to(device)
    tgt = val_batch["targets"].to(device)
    output = model(inpt).detach().flatten()
    hists = writeHistograms(inpt.T,output,tgt.flatten(),hists)
```

```python
plotsToSave = [plotFullEnergies, plotRatioOverCP]

for plotFct in plotsToSave:
    plotFct(hists)
    plt.savefig(os.path.join(model_path,plotFct.__name__ + ".png"))
```
